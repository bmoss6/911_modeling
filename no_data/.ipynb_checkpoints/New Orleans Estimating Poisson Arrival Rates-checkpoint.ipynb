```python
import pandas as pd
from scipy.stats import poisson
df = pd.read_csv("Calls_for_Service_2018.csv")
```

```python
df.index = pd.to_datetime(df['TimeCreate'])
day_watch = df.between_time('07:00','15:00')
swing_watch = df.between_time('15:00','23:00')
midnight_watch = df.between_time('23:00','07:00')
day_group = day_watch.groupby([day_watch.index.year, day_watch.index.month, day_watch.index.day, day_watch.index.hour])['NOPD_Item'].count()
swing_group = swing_watch.groupby([swing_watch.index.year, swing_watch.index.month, swing_watch.index.day, swing_watch.index.hour])['NOPD_Item'].count()
midnight_group = midnight_watch.groupby([midnight_watch.index.year, midnight_watch.index.month, midnight_watch.index.day, midnight_watch.index.hour])['NOPD_Item'].count()
all_group = df.groupby([df.index.year, df.index.month, df.index.day, df.index.hour])['NOPD_Item'].count()

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(3,1)
x = np.arange(len(day_group))
y = day_group.values
ax1.plot(x,y)

x = np.arange(len(swing_group))
y = swing_group.values
ax2.plot(x,y)

x = np.arange(len(midnight_group))
y = midnight_group.values
ax3.plot(x,y)
```

```python
from scipy.stats import poisson
from datetime import datetime
def plot_poisson_frequencies(group, time_slice, remove_outliers=False):
    p = poisson(group.mean())
    idx = group.idxmax()
    group = group.drop(labels=[idx])
    buckets = {x:0 for x in range(group.max()+1)}
    for x in group:
        buckets[x] += 1
    expected = {k: p.pmf(k)*len(group) for k in range(group.max()+1)}
    fig, ax = plt.subplots()
    ax.plot(buckets.keys(), buckets.values())
    ax.plot(expected.keys(),expected.values())
    ax.set_xlabel("# of Incidents Created in the New Orleans Computer Aided Dispatch System in an Hour from {}".format(time_slice))
    ax.set_ylabel("Frequency")
 
def plot_interarrival_frequencies(group,time_slice, remove_outliers=False):
    group = group.sort_index()
    inter_arrival_times = [group.index[x+1] - group.index[x] for x in len(group.index)]
#    buckets = {x:0 for x in range(group.max()+1)}
#    for x in group:
#        buckets[x] += 1
#    expected = {k: p.pmf(k)*len(group) for k in range(group.max()+1)}
#    fig, ax = plt.subplots()
#    ax.plot(buckets.keys(), buckets.values())
#    ax.plot(expected.keys(),expected.values())
#    ax.set_xlabel("# of Incidents Created in the New Orleans Computer Aided Dispatch System in an Hour from {}".format(time_slice))
#    ax.set_ylabel("Frequency")
plot_interarrival_frequencies(df, "All")
```

```python
plot_poisson_frequencies(day_group,'07:00-15:00')
plot_poisson_frequencies(swing_group, '15:00-23:00')
plot_poisson_frequencies(midnight_group, '23:00-07:00')
plot_poisson_frequencies(all_group, '00:00-24:00')
```
